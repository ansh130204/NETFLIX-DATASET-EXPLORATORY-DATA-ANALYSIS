Exploratory Data Analysis on Netflix Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix=pd.read_csv(r"C:\Users\ansha\Downloads\archive\netflix_titles.csv",index_col=0)
netflix
```

### Step 1: Dataset Information

```python
# Shape and info
print("Dataset Shape:", netflix.shape)
netflix.info()

# Check for missing values
netflix.isnull().sum().sort_values(ascending=False)

# Basic statistics
netflix.describe(include='all')
```

### Insights:

The dataset contains information about Netflix Movies and TV Shows.

There are missing values in columns like director and cast.

### Step 2: Data Cleaning

```python
# Remove duplicates
netflix.drop_duplicates(inplace=True)

# Handle missing values
netflix['country'].fillna('Unknown', inplace=True)
netflix['rating'].fillna('Not Rated', inplace=True)

# Convert date_added to datetime
netflix['date_added'] = pd.to_datetime(netflix['date_added'],errors='coerce')

# Extract year and month
netflix['year_added'] = netflix['date_added'].dt.year
netflix['month_added'] = netflix['date_added'].dt.month

netflix
```

## Step 3. Univariate Analysis
### 3.1 Type of Content

```python
sns.countplot(data=netflix, x='type', palette='Set2')
plt.title('Distribution of Movies and TV Shows on Netflix')
plt.xlabel('Type of Content')
plt.ylabel('Count')
plt.show()

netflix['type'].value_counts(normalize=True)*100
```

## Insights:

Netflixâ€™s catalog is dominated by Movies, forming about 70% of total titles.

TV Shows form a smaller segment but are growing gradually â€” reflecting Netflixâ€™s investment in series and original productions.

This indicates a strong focus on movie licensing and original TV content expansion.

## 3.2 Ratings Distribution

```python
sns.countplot(data=netflix, x='rating',
              order=netflix['rating'].value_counts().index, palette='viridis')
plt.title('Distribution of Ratings on Netflix')
plt.xticks(rotation=45)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

## Insights:

The most common rating is TV-MA, showing that Netflix largely targets mature audiences.

Other frequent ratings include TV-14 and R, further confirming the focus on adult and teenage content.

Childrenâ€™s content (TV-Y, TV-G) is less represented, suggesting Netflixâ€™s niche in general entertainment and mature drama rather than kidsâ€™ shows.

## Step 4. Country-wise Analysis

```python
top_countries = netflix['country'].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index, palette="flare")
plt.title('Top 10 Content-Producing Countries on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

## Insight:
The United States leads by a wide margin, followed by India, United Kingdom, Canada, and Japan.

This trend shows Netflixâ€™s American origin dominance, while also revealing significant growth in Asian content, particularly from India and Japan.

Indicates a global diversification strategy in Netflixâ€™s library.


## Step 5. Year-wise Additions

```python
netflix['year_added'].value_counts().sort_index().plot(kind='bar', color='salmon')
plt.title('Number of Titles Added Each Year')
plt.xlabel('Year Added')
plt.ylabel('Count')
plt.show()
```

## Insights:
A noticeable spike in content additions between 2016 and 2020 shows Netflixâ€™s massive expansion phase.

The platform added the highest volume of new content around 2019, coinciding with global expansion and increased original productions.

A small decline after 2020 likely reflects COVID-19-related production halts.


## Step 6. Genre Analysis

```python
genres = netflix['listed_in'].str.split(',').explode().str.strip().value_counts().head(15)

sns.barplot(x=genres.values, y=genres.index, palette="magma")
plt.title('Top 15 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```

## Insights:
International Movies, Dramas, and Comedies dominate Netflixâ€™s catalog, showing the platformâ€™s focus on globally appealing and story-driven content.

The strong presence of International TV Shows and Documentaries highlights Netflixâ€™s growing investment in diverse, cross-cultural, and factual entertainment.

A mix of Action, Romance, and Family genres reflects Netflixâ€™s effort to balance mainstream entertainment with content for varied audience preferences.

## Step 7. Duration Analysis (Movies Only)

```python
movies = netflix[netflix['type'] == 'Movie']
movies['duration'] = movies['duration'].str.replace(' min', '').astype(float)

sns.histplot(movies['duration'], bins=30, color='orange')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Count')
plt.show()
```

## Insights:
Most Netflix movies have a duration between 80 to 120 minutes, showing that the platform favors standard feature-length films.

The distribution is right-skewed, meaning very few movies exceed 150 minutes, as longer runtimes are less common among Netflix titles.

The concentration around 90â€“100 minutes indicates Netflixâ€™s preference for compact, easily consumable movies suited for a broad streaming audience.
